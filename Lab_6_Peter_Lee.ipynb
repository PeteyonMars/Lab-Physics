```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import glob
import os
import pandas as pd
import csv
```

```python
'''
Import data
'''

path = "/content/drive/MyDrive/Lab Physics/G Lab data"

def load_data(folder_path):
  data_dict = {}

  files = glob.glob(os.path.join(folder_path, "*.csv"))
  print(f"Reading {folder_path}: {files}")

  for file in files:
    try:
      file_data = np.genfromtxt(file, delimiter = ",", skip_header = 1)

      time = file_data[:, 0]
      position = file_data[:, 1]

      filename = os.path.basename(file)
      data_dict[filename] = (time, position)

    except Exception as e:
      print(f"Error for {file}: {e}")

  return data_dict

G_data = load_data(path)
```

```python
'''
Graph and fit damped oscillator equation
'''

pi = np.pi
def equation(a, A, w, t, phi, B): # a for decay constant, A for amplitude, w for rotational frequency, t for independent variable, phi for phase offset, and B for vertical offset
  return A * np.exp(-a * t) * np.cos(w * t + phi) + B

a_vals = []
w_vals = []
def graph_data(data):
  for file,(time, position) in data.items():
    mask = position < 5000
    x_data = time[mask]

    y_data = position[mask]
    centered_y_data = y_data

#Peak Finding

    prom = 0.15
    height = -1
    dist = 150

    peaks, properties = sp.signal.find_peaks(y_data, prominence = prom, height = height, distance = dist)
    print("========\nIndices of the peaks:", peaks)
    #print("Properties:", properties)
    x_peaks = x_data[peaks]
    y_peaks = y_data[peaks]

    #if len(x_peaks) >= 2:
    #n_peaks = len(x_peaks)
    #initial = 0
    #final = n_peaks - 1
    #avg_T = (x_peaks[final] - x_peaks[initial]) / n_peaks
    #avg_T_vals.append(avg_T)

# Fitting
    t = x_data
    y = centered_y_data
    params = Parameters()

    params.add('a', vary = True, value = 7.7995e-04, min = 0.0005, max = 0.005)
    params.add('A', vary = True, value = 4, min = 1, max = 5)
    params.add('w', vary = True, value = 0.001, min = 0.01, max = 0.05)
    params.add('phi', vary = True, value = pi / 2, min = -2 * pi, max = 2 * pi)
    params.add('B', vary = True, value = 0)

    model = Model(equation, independent_vars=['t']) # Defining t as the independent variable

    result = model.fit(y, t = t, params = params)


    a_val = result.params['a'].value

    #a_vals.append(a_val)

    A_val = result.params['A'].value

    w_val = result.params['w'].value
    #w_vals.append(w_val)
    if A_val > 1.5:
      a_vals.append(a_val)
      w_vals.append(w_val)

    print(result.fit_report())

    plt.figure(figsize=(5,3))
    plt.title(f"Graph for {file}", fontsize = 14)
    #plt.plot(x_peaks, y_peaks, 'r*')
    plt.plot(x_data, y_data, 'b', label = "Raw Plot V vs t")
    plt.plot(t, result.best_fit, 'g', label = "Best Fit")
    plt.xlabel("Time (s)", fontsize = 14)
    plt.ylabel("Position (V)", fontsize = 14)
    plt.legend(fontsize = 10)
    plt.show()

graph_data(G_data)

print(f"a vals: {a_vals}")
print(f"w vals: {w_vals}")

'''
Find G value
'''
d = 0.061 # distance the laser covers measured by Howard
L = 2.06 # pendulum to laser
H = 0.05 # axis perpendicular to the axis of rotation to big orb
M = 1.503 # mass of the big orb
l = 0.0405 # radius of the rod between the two small orbs

# What to do with d?

#w_vals = []
#def find_constants():
  #theta = d / (4 * L) # from board
  #for w in w_vals:
    #w_val = (2 * pi) / T
    #w_vals.append(w_val)

#find_constants()
print(f"w vals: {w_vals}")

G_vals = []
def find_G():
  for i in range(len(w_vals)):
    a = a_vals[i]
    w = w_vals[i]
    G_val = ((l * d * (w ** 2 + a ** 2)) / (4 * M * L)) * (H - l * (d / (4 * L))) ** 2
    G_vals.append(G_val)

find_G()
print(f"G vals: {G_vals}")
print(len(G_vals))

def final_values():
  avg_G = np.mean(G_vals)
  N = len(G_vals)
  S = np.std(G_vals)

  u = S / np.sqrt(N)

  return avg_G, u

avg_G, u = final_values()
print(f"Final G measurement: ({avg_G} ± {u}) N * m^2 / kg^2")

accepted_G = 6.6743e-11
def percent_error():
  error = np.abs((avg_G - accepted_G) / accepted_G) * 100
  return error

error = percent_error()
print(f"Percent error: {error} %")
```

```python
G_actual = accepted_G
G_list = list(G_vals)
x = list(range(len(G_list)))
y = G_list

u_G = [u] * len(G_list)

fig=plt.figure()
plt.errorbar(x, y, yerr = u_G, linestyle = "-", marker = "o", markersize = 3, c = 'black', elinewidth = 1, capsize = 3, label = "Uncertainty")
plt.axhline(y = G_actual, color = 'g', linestyle = "--", label = "Accepted G Value")
plt.title("Uncertainty Graph for G Measurements", fontsize = 14)
plt.xlabel("N", fontsize = 12)
plt.ylabel("G (N * m^2 / kg^2)", fontsize = 12)
plt.legend(fontsize = 12)

plt.grid()
```
