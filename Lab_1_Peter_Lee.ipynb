```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import pandas as pd
```

```python
path="/content/drive/MyDrive/Lab Physics/Lab_#1_Air_Data.csv"
read=pd.read_csv(path)
data=read.to_numpy()

n1_f1=data[:, 0]
n1_f2=data[:, 1]
n1_f3=data[:, 2]
n2_f1=data[:, 3]
n2_f2=data[:, 4]
n2_f3=data[:, 5]
n3_f1=data[:, 6]
n3_f2=data[:, 7]
n3_f3=data[:, 8]
n4_f1=data[:, 9]
n4_f2=data[:, 10]
n4_f3=data[:, 11]
n7_f1=data[:, 12]
n7_f2=data[:, 13]
n7_f3=data[:, 14]

#Making sure the import was correct
print(n1_f1)
print(n1_f2)
print(n1_f3)

#Number of nodes a box had for a frequency, so n1 is box 1, n2 is box 2, etc.
n1=[n1_f1, n1_f2, n1_f3]
n2=[n2_f1, n2_f2, n2_f3]
n3=[n3_f1, n3_f2, n3_f3]
n4=[n4_f1, n4_f2, n4_f3]
n7=[n7_f1, n7_f2, n7_f3]

#Uncertainty values already calculated in lab notebook
uf=0.001
uLx=0.00003
uLy=0.00003
uLz=0.00003

#First=x, second=y, third=z
#The measurments for a box's dimensions.
L1=np.array([0.164, 0.117, 0.027])
L2=np.array([0.119,0.109,0.025])
L3=np.array([0.163,0.119,0.075])
L4=np.array([0.136,0.067,0.099])
L7=np.array([0.13483,0.09874,0.01981])

#First=f1, Second=f2, Third=f3
f=np.array([2314,4199,6272])
f1=f[0]
f2=f[1]
f3=f[2]

pi=np.pi

def calculate_error(f, n, L, box_number):
  nx,ny,nz=n
  Lx,Ly,Lz=L

  dv_df=2 * np.pi / (np.sqrt((nx / Lx)**2 + (ny / Ly)**2 + (nz / Lz)**2))
  dv_dLx=2 * np.pi * f * (nx)**2 / (((Lx)**3) * (((nx)**2/(Lx)**2) + ((ny)**2/(Ly)**2) + ((nz)**2/(Lz)**2))**(3/2))
  dv_dLy=2 * np.pi * f * (ny)**2 / (((Lx)**3) * (((nx)**2/(Lx)**2) + ((ny)**2/(Ly)**2) + ((nz)**2/(Lz)**2))**(3/2))
  dv_dLz=2 * np.pi * f * (nz)**2 / (((Lx)**3) * (((nx)**2/(Lx)**2) + ((ny)**2/(Ly)**2) + ((nz)**2/(Lz)**2))**(3/2))
  Error_v=np.sqrt((dv_df**2) * (uf**2) + (dv_dLx**2) * (uLx**2) + (dv_dLy**2) * (uLy**2) + (dv_dLz**2) * (uLz**2))
  print(f"The calculation for box {box_number} at {f} Hz in air was {Error_v}")

calculate_error(f1, n1_f1, L1, 1)
calculate_error(f2, n1_f2, L1, 1)
calculate_error(f3, n1_f3, L1, 1)
print()
calculate_error(f1, n2_f1, L2, 2)
calculate_error(f2, n2_f2, L2, 2)
calculate_error(f3, n2_f3, L2, 2)
print()
calculate_error(f1, n3_f1, L3, 3)
calculate_error(f2, n3_f2, L3, 3)
calculate_error(f3, n3_f3, L3, 3)
print()
calculate_error(f1, n4_f1, L4, 4)
calculate_error(f2, n4_f2, L4, 4)
calculate_error(f3, n4_f3, L4, 4)
print()
calculate_error(f1, n7_f1, L7, 7)
calculate_error(f2, n7_f2, L7, 7)
calculate_error(f3, n7_f3, L7, 7)
print()

print("From the values of speed of sound in air from Google Sheets,")
v_box1_f1=310.684
print(f"Speed of sound in air at {f1} Hz for box 1 is {v_box1_f1} m/s")
v_box1_f2=324.517
print(f"Speed of sound in air at {f2} Hz for box 1 is {v_box1_f2} m/s")
v_box1_f3=314.924
print(f"Speed of sound in air at {f3} Hz for box 1 is {v_box1_f3} m/s")
print()
v_box2_f1=372.729
print(f"Speed of sound in air at {f1} Hz for box 2 is {v_box2_f1} m/s")
v_box2_f2=338.178
print(f"Speed of sound in air at {f2} Hz for box 2 is {v_box2_f2} m/s")
v_box2_f3=390.015
print(f"Speed of sound in air at {f3} Hz for box 2 is {v_box2_f3} m/s")
print()
v_box3_f1=377.992
print(f"Speed of sound in air at {f1} Hz for box 3 is {v_box3_f1} m/s")
v_box3_f2=336.885
print(f"Speed of sound in air at {f2} Hz for box 3 is {v_box3_f2} m/s")
v_box3_f3=333.158
print(f"Speed of sound in air at {f3} Hz for box 3 is {v_box3_f3} m/s")
print()
v_box4_f1=370.532
print(f"Speed of sound in air at {f1} Hz for box 4 is {v_box4_f1} m/s")
v_box4_f2=315.738
print(f"Speed of sound in air at {f2} Hz for box 4 is {v_box4_f2} m/s")
v_box4_f3=338.316
print(f"Speed of sound in air at {f3} Hz for box 4 is {v_box4_f3} m/s")
print()
v_box7_f1=351.949
print(f"Speed of sound in air at {f1} Hz for box 7 is {v_box7_f1} m/s")
v_box7_f2=346.114
print(f"Speed of sound in air at {f2} Hz for box 7 is {v_box7_f2} m/s")
v_box7_f3=345.981
print(f"Speed of sound in air at {f3} Hz for box 7 is {v_box7_f3} m/s")
print()
print("If we were interested in finding the speed of sound with its uncertainty, we would bring a speed with its corresponding uncertainty:")
print()
print(f"e.g., the speed of sound in air for box 1 at {f1} Hz in air was 310.7±0.1 m/s")
print("And etc.")
print()
print("But we are interested in finding the speed of sound, and I think the best way of doing that is to use a weighted average by combining the speed values with its uncertainty, of which")
print("the latter was found using propagation error analysis, which shows how much each variable contributes to the overall error of a final value, which in this case is speed.")

v_values=np.array([310.684, 324.517, 314.924, 372.729, 338.178, 390.015, 377.992, 336.885, 333.158, 370.532, 315.738, 338.316, 351.949, 346.114, 345.981])
u_values=np.array([0.12345219598842351, 0.1666406041926899, 0.11264864414610054, 0.19008550472875074, 0.17246479533304995, 0.2060232200851869, 0.2180899631996146, 0.11625643933520816, 0.09423755192519265, 0.12574129842600998, 0.15081311227117863, 0.09052417855214091, 0.12722889013009764, 0.16804428397130672, 0.1827372729989255])

def findBest(v_values, u_values):
  weight=1/(u_values**2)
  v_best=np.sum(v_values*weight)/np.sum(weight)
  u_best=np.sqrt(1/np.sum(weight))
  print(f"The final speed of sound in air is {v_best:.2f}±{u_best:.2f} m/s")

findBest(v_values, u_values)

#Uncertainty graph
x_axis = np.arange(0, len(v_values), step=1)
y1=v_values
y2=u_values
fig=plt.figure()
plt.errorbar(x_axis, y1, yerr = y2, linestyle = "-", marker = "o", markersize = 3, c = 'r', elinewidth = 1, capsize = 3, label = "uncertainty")
plt.title("Air Uncertainty Graph", fontsize = 14)
plt.xlabel("N", fontsize = 12)
plt.ylabel("", fontsize = 12)
plt.legend(fontsize = 12)
plt.grid()
```
